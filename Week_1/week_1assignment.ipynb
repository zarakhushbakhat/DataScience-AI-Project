```python
# Import libraries
import pandas as pd
import seaborn as sns
from tabulate import tabulate

# Optional: expand display for full view
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 200)

# Load Titanic dataset
titanic = sns.load_dataset("titanic")

# Column descriptions
column_info = {
    "survived": "0 = did not survive, 1 = survived",
    "pclass": "Passenger class (1 = 1st, 2 = 2nd, 3 = 3rd)",
    "sex": "Gender (male/female)",
    "age": "Passenger age in years",
    "sibsp": "Number of siblings/spouses aboard",
    "parch": "Number of parents/children aboard",
    "fare": "Ticket price paid",
    "embarked": "Port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton)",
    "class": "Passenger class as text (First, Second, Third)",
    "who": "Person type (man, woman, child)",
    "adult_male": "True if adult male, False otherwise",
    "deck": "Deck level (A, B, C, etc.)",
    "embark_town": "Boarding city",
    "alive": "Yes if survived, No if not",
    "alone": "True if traveling alone"
}

# Print column descriptions
print("ðŸ“Œ Column Descriptions:")
for col, desc in column_info.items():
    print(f"{col}: {desc}")

# Display first 10 rows in formatted table
print("\nðŸ“Œ Titanic Dataset Sample:")
print(tabulate(titanic.head(10), headers="keys", tablefmt="grid"))
```
